```python
#Carga de las librerias
import openai
import gradio as gr
import os
import docx
```

```python
#Llave para acceder a la API de OpenAI
openai.api_key =""
```

```python
#Funciónes para interrogar los documentos
def obtener_valor_del_contrato(texto):
    prompt = f"Del siguiente texto del contrato, necesito saber exclusivamente el valor total del contrato expresado en cifras. Por favor, responde únicamente con la cantidad numérica sin símbolos de moneda y palabras, o cualquier otro carácter adicional. Ejemplo: si el valor del contrato es de diez mil dólares, responde '10000'. Aquí está el texto del contrato: {texto}"
    #prompt = f"Del siguiente contrato: {texto} ¿Cuál es el valor total del contrato?, unicamente vas a responer con el número: ejemplo 10.000"
    #prompt = f"Del siguiente texto del contrato: {texto}¿Cuál es el valor total del contrato?, incluyendo puntos para separar los miles y comas para los decimales. Por ejemplo, si el valor es de noventa y cinco millones trescientos quince mil trescientos ochenta y cuatro, responde '95.315.384,00'. Por favor, proporciona solo la cifra con el formato correcto. sin símbolos de moneda y palabras"

    respuesta = openai.ChatCompletion.create(
        model="gpt-4-1106-preview",
        messages=[
            {"role": "system", "content": "Estás hablando con un asistente que puede analizar contratos y determinar el valor del contrato y responder solo con el número."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=12
    )
    return respuesta['choices'][0]['message']['content']

def obtener_contratante(texto):
    #prompt = f"Del siguiente contrato: {texto} ¿Quíen es el contratista? Responde solo con el nombre del contratista."
    prompt = f"Revisando el contenido de este contrato: {texto} Por favor, identifica y proporciona exlusivamente el nombre del contratante. Si por ejemplo el contratante es 'Construcciones S.A.', simplemente responde 'Construcciones S.A.'."
    respuesta = openai.ChatCompletion.create(
        model="gpt-4-1106-preview",
        messages=[
            {"role": "system", "content": "Estás hablando con un asistente que puede identificar entidades en contratos y responder solo con el nombré de las entidades."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=30
    )
    return respuesta['choices'][0]['message']['content']


def obtener_fecha_contrato(texto):
    #prompt = f"Analiza el siguiente contrato y proporciona solo la fecha del contrato: {texto}"
    prompt = (
    "Por favor, revisa el siguiente texto del contrato y encuentra la fecha en la que fue firmado o entró en vigor. "
    "Responde solamente con la fecha en formato DD/MM/AAAA. No incluyas ninguna otra información. "
    "Por ejemplo, si el contrato fue firmado el primero de enero de 2023, simplemente responde '01/01/2023'. "
    f"Aquí está el texto del contrato: {texto}"
    )
    respuesta = openai.ChatCompletion.create(
        model="gpt-4-1106-preview",
        messages=[
            {"role": "system", "content": "Estás hablando con un asistente que puede determinar fechas de contratos."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=10
    )
    return respuesta['choices'][0]['message']['content']

def obtener_ciudad_contrato(texto):
    #prompt = f"Analiza el siguiente contrato y proporciona solo la ciudad relacionada {texto}"
    prompt = f"Del siguiente texto de contrato, necesito saber únicamente la ciudad en la que se aplica o se firma el contrato. Por favor, indica solo el nombre de la ciudad de forma clara y concisa. Ejemplo: Si el contrato se firma en 'Bogotá', simplemente responde 'Bogotá'. Aquí está el texto del contrato: {texto}"

    respuesta = openai.ChatCompletion.create(
        model="gpt-4-1106-preview",
        messages=[
            {"role": "system", "content": "Estás hablando con un asistente que puede identificar ciudades en contratos."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=15
    )
    return respuesta['choices'][0]['message']['content']
def obtener_objeto_contrato(texto):
    #prompt = f"Por favor, analiza el siguiente contrato y proporciona una descripción clara y concisa del objeto del contrato. El objeto del contrato se refiere a la principal finalidad o propósito del acuerdo. Asegúrate de incluir detalles relevantes sobre lo que implica el contrato. Aquí está el texto del contrato: {texto}"
    prompt = f"Por favor,analiza el siguiente contrato y proporciona exclusivamente el objeto del contrato: {texto}"
    respuesta = openai.ChatCompletion.create(
        model="gpt-4-1106-preview",
        messages=[
            {"role": "system", "content": "Estás hablando con un asistente que puede analizar contratos y determinar el objeto del contrato."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=132  # Ajusta este valor según la longitud deseada de la respuesta
    )
    return respuesta['choices'][0]['message']['content']
```

```python
#Función para leer el archivo en formato Docx, adémas combinas las funciones que interrogan a los documentos
def combinar_info_contrato_1(archivo):
    # Utiliza 'docx' para abrir el archivo de Word
    doc = docx.Document(archivo)
    texto = ' '.join([p.text for p in doc.paragraphs])

    # Obtiene la información requerida
    valor_contrato = obtener_valor_del_contrato(texto)
    contratante = obtener_contratante(texto)
    fecha = obtener_fecha_contrato(texto)
    ciudad = obtener_ciudad_contrato(texto)
    objeto = obtener_objeto_contrato(texto)
    nombre_archivo = os.path.basename(archivo)  # Esto extrae el nombre del archivo de la ruta completa
    contrato = os.path.splitext(nombre_archivo)[0]  # Esto elimina la extensión del archivo
    # Crea un diccionario con la información
    info_contrato = {
        "Archivo": contrato,
        "Fecha": fecha,
        "Ciudad": ciudad,
        "Valor del contrato": valor_contrato,
        "Contratante": contratante,
        "Objeto del contrato": objeto
        #"Tipo de contrato": tipo_contrato
    }

    # Devuelve el diccionario directamente
    return info_contrato
```

```python
def gradio_interface(uploaded_file):
    # Gradio pasa un objeto de archivo temporal a la función
    return combinar_info_contrato_1(uploaded_file)

# Define la interfaz de Gradio
iface = gr.Interface(
    fn=gradio_interface,
    inputs=gr.File(label="Sube tu contrato en formato docx aquí"),
    outputs="json"

)

# Lanza la interfaz de Gradio
iface.launch()
```
